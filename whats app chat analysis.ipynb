```python
# function to split the year , month and the year
def years(text):
    return text.split('/')[2]

def month(text):
    return text.split('/')[1]

def days(text):
    return text.split('/')[0]
```

```python
# extract all the days from the dtaaframes 
df['year']=df['dates'].apply(years)
df['month']=df['dates'].apply(month)
df['day']=df['dates'].apply(days)
```

```python
# to see how many words are spoken b y the individuals users 
def detail_of_contributions(name):
    temp_df=df[df['User Name']==name]
    print(f"**** Detail of {name}  **************\n")
    print("  Total Message : ",temp_df.shape[0])
    temp_words=[]
    for message in temp_df['Messages']:
        temp_words.extend(message.split())
    print("  Total Word :",len(temp_words))
    print("  Overall message Percentage :",np.round((temp_df.shape[0]/df.shape[0])*100,2)," %")
    print("  Overall word percentages : ",np.round((len(temp_words)/len(words)*100),2)," %")
    print("  Overall Media shared :",len(temp_df[temp_df['Messages']=='<Media omitted>']))
```

```python
# make the word cloud of the group chats 
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# Generate word cloud
wc = WordCloud(width=300, height=300, min_font_size=10, background_color='white')
wordcloud = wc.generate(' '.join(df['Messages']))

# Display the word cloud
plt.figure(figsize=(7, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# make tyhe wordcloud based on the user names
def word_stu(name):
    temp_df=df[df['User Name']==name]
    wc = WordCloud(width=300, height=300, min_font_size=10, background_color='white')
    wordcloud = wc.generate(' '.join(temp_df['Messages']))

    # Display the word cloud
    plt.figure(figsize=(7, 10))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    word_frequencies = wordcloud.words_
    
    print("\n")
    print(word_frequencies)
    print(type(word_frequencies))
    
```
